```python

import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# HH parametreleri
C_m   = 1.0
g_Na  = 120.0
g_K   = 36.0
g_L   = 0.3
E_Na  = 50.0
E_K   = -77.0
E_L   = -54.4

def alpha_m(V): return 0.1 * (V + 40.0) / (1.0 - np.exp(-(V + 40.0) / 10.0))
def beta_m(V):  return 4.0 * np.exp(-(V + 65.0) / 18.0)
def alpha_h(V): return 0.07 * np.exp(-(V + 65.0) / 20.0)
def beta_h(V):  return 1.0 / (1.0 + np.exp(-(V + 35.0) / 10.0))
def alpha_n(V): return 0.01 * (V + 55.0) / (1.0 - np.exp(-(V + 55.0) / 10.0))
def beta_n(V):  return 0.125 * np.exp(-(V + 65.0) / 80.0)

def hh_deriv(state, t, I_ext):
    V, m, h, n = state
    dVdt = (I_ext - g_Na * m**3 * h * (V - E_Na) - g_K * n**4 * (V - E_K) - g_L * (V - E_L)) / C_m
    dmdt = alpha_m(V) * (1.0 - m) - beta_m(V) * m
    dhdt = alpha_h(V) * (1.0 - h) - beta_h(V) * h
    dndt = alpha_n(V) * (1.0 - n) - beta_n(V) * n
    return [dVdt, dmdt, dhdt, dndt]

class EBESM_v7:
    def __init__(self, num_neurons=100, avg_degree=10, rewire_prob=0.1, seed=42):
        self.N = num_neurons
        self.graph = nx.watts_strogatz_graph(self.N, avg_degree, rewire_prob, seed=seed)
        self.coupling = 0.08

        self.states = np.zeros((self.N, 4))
        self.states[:,0] = -65.0 + np.random.randn(self.N)*5
        V0 = self.states[:,0]
        self.states[:,1] = alpha_m(V0)/(alpha_m(V0)+beta_m(V0))
        self.states[:,2] = 0.6
        self.states[:,3] = 0.32

        self.time = np.linspace(0, 300, 2000)  # daha hızlı simülasyon

    def synaptic_input(self, i, voltages):
        neighbors = list(self.graph.neighbors(i))
        if not neighbors:
            return 0.0
        return self.coupling * np.mean(voltages[neighbors] - voltages[i])

    def network_deriv(self, flat_state, t):
        states = flat_state.reshape((self.N,4))
        voltages = states[:,0]
        deriv = np.zeros_like(states)
        for i in range(self.N):
            I_ext = 9.0
            I_syn = self.synaptic_input(i, voltages)
            deriv[i] = hh_deriv(states[i], t, I_ext+I_syn)
        return deriv.flatten()

    def simulate(self):
        print("EBESM v7 simülasyonu başlıyor...")
        sol = odeint(self.network_deriv, self.states.flatten(), self.time)
        self.voltages = sol.reshape((len(self.time), self.N, 4))[:,:,0]
        print("Simülasyon tamamlandı.")

    def synchronization_index(self):
        mean_trace = np.mean(self.voltages, axis=1)
        R = np.std(mean_trace)/np.mean(np.std(self.voltages, axis=1))
        return R

    def plot_raster(self, num_sample=40):
        sample_idx = np.random.choice(self.N, num_sample, replace=False)
        plt.figure(figsize=(10,6))
        for i, idx in enumerate(sample_idx):
            plt.plot(self.time, self.voltages[:,idx]+i*60, color='black', lw=0.8)
        plt.title(f"EBESM v7 - Spiking Raster Plot ({num_sample} örnek)")
        plt.xlabel("Zaman (ms)")
        plt.ylabel("Nöron ID (offset ile)")
        plt.xlim(50,300)
        plt.show()

    def plot_network(self):
        plt.figure(figsize=(6,6))
        nx.draw_circular(self.graph, node_size=20, node_color='skyblue', edge_color='gray', alpha=0.6)
        plt.title("EBESM v7 - Küçük Dünya Nöron Ağı")
        plt.show()

# ========================
if __name__ == "__main__":
    model = EBESM_v7()
    model.simulate()

    sync_index = model.synchronization_index()
    print(f"\nEmergent Senkronizasyon İndeksi (Bilinç Proxy'si): {sync_index:.4f}")

    model.plot_raster()
    model.plot_network()
```
